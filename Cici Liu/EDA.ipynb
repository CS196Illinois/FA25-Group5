```python
#filter by time
from datetime import time 

#convert text to datetime, HH:MM format
#df['Start Time Clean'] = pd.to_datetime(df['Start Time'], format='%H:%M', errors='coerce')
#df['End Time Clean'] = pd.to_datetime(df['End Time'], format='%H:%M', errors='coerce')
df['Start Time'] = pd.to_datetime(df.iloc[:, 20], format='%I:%M%p', errors='coerce')  
df['End Time'] = pd.to_datetime(df.iloc[:, 21], format='%I:%M%p', errors='coerce')
df = df.dropna(subset=['Start Time', 'End Time'])


#check if conversion worked
print(df['Start Time'])
print(df['End Time'])
print(df.iloc[:, 20].head(20))  # Start Time (column U)
print(df.iloc[:, 21].head(20))  # End Time (column V)

#define filter times
start_filter = time(10,0) 
end_filter = time(18,0)

#filter rows
filtered_df = df[(df['Start Time'].dt.time >= start_filter) & (df['End Time'].dt.time <= end_filter)]


filtered_df

```
