```python
import pandas as pd
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/CS196Illinois/FA25-Group5/refs/heads/master/Yaoxin%20Jiang/modified-2025-sp%20(with%20RMP).csv")
df
```

```python
df["Primary Instructor"].dropna()
```

```python
def calculate_log21(n, m, weight_excellent, weight_oustanding):
    if np.isnan(n) or np.isnan(m):
        return -1;
    value = n * weight_excellent + m * weight_oustanding + 1
    result = np.log(value) / np.log(21)  # log base 21
    return result
weight_excellent = 0.7
weight_oustanding = 1
df["log21_result"] = df.apply(
    lambda row: calculate_log21(row["Excellents"], row["Outstandings"], weight_excellent, weight_oustanding), 
    axis=1
)
df
```

```python
def calculate_RMP(n):
    return n * 20
df["RMP Score"] = df.apply(
    lambda row: calculate_RMP(row["RMP"]), 
    axis=1
)
df
```
